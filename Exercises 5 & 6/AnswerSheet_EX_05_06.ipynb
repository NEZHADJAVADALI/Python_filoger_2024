    <p>
    Create a Python program that first prompts the user to enter 10 numbers, which are then stored in a list. Subsequently, request the user to input a new number. If this new number is present in the list, remove all occurrences of it. if it1s not found, insert the number twice at the beginning of the list. Finally, display the modified list to the user.
    </p>

```python
# get 10 number from user and store in a list
num_list = []
for i in range(1,11):
    num_list.append(int(input("Please Enter 10 number: ")))

# take the check number from user
check_number = int(input("Please Enter a check number: "))

# check if the check number is in num_list
if check_number in num_list:
    num_list.remove(check_number)
else:
    num_list = [check_number,check_number] + num_list
        
print(num_list)
```
